## Cleaning and Preparing the data for model training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Run this once at the top of your notebook to stop all downcasting warnings
pd.set_option('future.no_silent_downcasting', True)
```

```python
#import the dataset
df_train = pd.read_csv('blackfriday.csv')
df_train.head()
```

```python
df_train.shape
```

```python
#import the dataset
df_test = pd.read_csv('blackfriday_test.csv')
df_test.head()
```

```python
df_test.shape
```

```python
#Merge Both train and test data
df = pd.concat([df_train,df_test])
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
df.head()
```

```python
## Handling categorical feature Gender
df['Gender']=df['Gender'].replace({'F': 0,'M': 1}).infer_objects(copy=False)
df.head()
```

```python
df['Age'].unique()
```

```python
## Handling categorical feature Age
df['Age']=df['Age'].replace({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
## fixing categorical feature City_Category
df_citycateg = pd.get_dummies(df['City_Category'],drop_first=True,dtype=int)
df_citycateg.head()
```

```python
df=pd.concat([df,df_citycateg],axis=1)
df.head()
```

```python
## dropping City_Category feature as its encoded
df.drop('City_Category',axis=1,inplace=True)
df.head()
```

```python
## Missing values
df.isnull().sum()
```

```python
## Focus on replacing missing values
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
## Mode
df['Product_Category_2'].mode()
```

```python
df['Product_Category_2'].mode()[0].item()
```

```python
## Replace the missing values with mode
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum().item()
```

```python
## Replace Product_Category_3 Missing values
df['Product_Category_3'].unique()
```

```python
df['Product_Category_3'].value_counts()
```
