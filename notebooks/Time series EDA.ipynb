## time Resampling

```python
df_tesla.head()
```

```python
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()
```

```python
##Year end frequency (rule='A')

df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
## Quarterly Start Frequency (QS)

df_tesla.resample(rule='QS').max()
```

```python
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
df_tesla['Open'].resample('M').mean()
```

```python
df_tesla['Open'].resample('M').mean().plot(kind='bar',figsize=(15,6))
```
